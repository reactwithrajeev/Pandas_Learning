# Apply Functions

```python
def age_category (age):
    if age <=24:
        return "Students"
    elif (age>24) and (age<50):
        return 'Adults'
    else :
        return 'Aged'
```

```python
ins_data['age_category'] = ins_data['age'].apply(age_category)
```

```python
ins_data['age_category'].value_counts()
```
